Q4 - The folder Stock_Data contains stock price information (open, hi, low, close, adj close, volume) on all of the stocks listed in stock_tickers.csv.  For each of the stocks listed in this file, we would like to compute the average open price for the first quarter and write these results to new csv called Q1_Results.csv.


a)First read the 20 stock tickers into a list from the file stock_tickers.csv

```python
#Write your code here

file=open("Stock_Data/Stock_tickers.csv","r")
content=file.readlines()


list_stickers=[]

for line in content:
    list_stickers.append(line.strip("\n"))
    
print(list_stickers)
```

   b) Next, create a dictionary where there is a key for each stock and the values are a list of the opening prices for the first quarter

```python
#Write your code here
stockopen={}

for stock in list_stickers:
    priceopen=[]
    file=open(f"Stock_Data/{stock}.csv","r")
    content=file.readlines()
    for price in content[1:63]:
        pricecal=price.split(",")
        priceopen.append(float(pricecal[2]))
    stockopen[stock]=priceopen
    
print(stockopen)
```

c)The final step is writing the result to a new csv called Q1_results.csv

```python
#Write your code here
file=open("Stock_Data/Q1_results.csv","w")
file.write("stock,average open price\n")
for stock in stockopen:
    stockprice=stockopen[stock]
    stockavg=sum(stockprice)/len(stockprice)
    file.write(f"{stock},{stockavg:.2f}\n")
f.close()

f = open("Stock_Data/Q1_results.csv")
lines = f.readlines()
print(lines)
```
