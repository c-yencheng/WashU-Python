We will write code to answer the following questions:

1. What fraction of customers order burritos versus burrito bowls?
2. What is the average amount of money spent on a Chipotle order?

For question 1, we will use the apply method to create a column that will read "Burrito" for items that are a burrito, "Bowl" for items that are a bowl and "Other" for every other item.

```python
#Write your code here

def readchipotle(row):
    
    eat=row['item_name'].lower()
    
    if "burrito" in eat:
        return "Burrito"
    elif "bowl" in eat:
        return "Bowl"
    else:
        return "Other"
        
```

Create new column called "Bowl_Burrito" and compute fractions of bowl versus burrito orders. The value_counts() method could be useful here.

```python
#Write you code here
df_chipotle["Bowl_Burrito"] = df_chipotle.apply(readchipotle, axis=1)
df_chipotle.head()
df_chipotle['Bowl_Burrito']
```

```python
fraction=df_chipotle['Bowl_Burrito'].value_counts().Bowl/df_chipotle['Bowl_Burrito'].value_counts().Burrito
print(f'The fraction is about {fraction:.2f}')
```

For question 2, use iterrows to calculate the average price per order. Note that the item price column is a string, so first use string methods plut the as.type() method to recast this column as a float.

```python
df_chipotle.shape[0]
df_chipotle['item_price']
```

```python
#Write your code here

#Convert item_price to float
df_chipotle['item_price']=df_chipotle['item_price'].str.strip('$').astype('float64')
df_chipotle.dtypes
```

```python
D={}
for index,row in df_chipotle.iterrows():
    
    pricefloat=row['item_price']
    try:
        D[row['order_id']]+=pricefloat/df_chipotle.shape[0]
    except KeyError:
        D[row['order_id']]=pricefloat/df_chipotle.shape[0]

D
```
